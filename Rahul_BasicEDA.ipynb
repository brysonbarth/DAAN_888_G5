```python
from kaggle.api.kaggle_api_extended import KaggleApi
import pandas as pd
api = KaggleApi()
api.authenticate()
api.dataset_download_file('thedevastator/cancer-patients-and-air-pollution-a-new-link','cancer patient data sets.csv', path='./')
df = pd.read_csv('cancer%20patient%20data%20sets.csv')
```

```python
df
```

```python
df_rahul = df[['Age','Shortness of Breath','Wheezing','Swallowing Difficulty', 'Clubbing of Finger Nails']]
df_rahul
```

```python
df_rahul.isnull().values.any()
```

```python
df_rahulmin = df_rahul.min()
df_rahulmin
```

```python
df_rahulmax = df_rahul.max()
df_rahulmax
```

```python
df_rahul.dtypes
```

```python
def mapping(risk):
    if risk <= 3:
        return "low"
    if risk <= 6:
        return "medium"
    if risk <= 9:
        return "high"

df_rahul["Shortness of Breath"] = df_rahul["Shortness of Breath"].map(mapping)
df_rahul["Wheezing"] = df_rahul["Wheezing"].map(mapping)
df_rahul["Swallowing Difficulty"] = df_rahul["Swallowing Difficulty"].map(mapping)
df_rahul["Clubbing of Finger Nails"] = df_rahul["Clubbing of Finger Nails"].map(mapping)
```

```python
df_rahul
```

```python
df_rahul['Binned Age'] = pd.cut(df_rahul['Age'], 3)
df_rahul
```

```python
df_shortbreath = df_rahul.groupby('Binned Age')['Shortness of Breath'].value_counts().unstack()
df_wheez = df_rahul.groupby('Binned Age')['Wheezing'].value_counts().unstack()
df_swaldiff = df_rahul.groupby('Binned Age')['Swallowing Difficulty'].value_counts().unstack()
df_cfingnail = df_rahul.groupby('Binned Age')['Clubbing of Finger Nails'].value_counts().unstack()
```

```python
df_shortbreath.plot(kind='bar', stacked=True)
```
